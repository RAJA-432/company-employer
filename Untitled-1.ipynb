```python
# Query all employees and display in a clean table format
all_employees = employees_collection.all()  # TinyDB all() returns list of dicts
df = pd.DataFrame(all_employees)
print("Employee Database:")
print(df.to_string(index=False))
```

```python
# Example query: Find employees in Engineering department
q = Query()
engineering_employees = employees_collection.search(q.department == "Engineering")
df_eng = pd.DataFrame(engineering_employees)
print("\nEngineering Department Employees:")
print(df_eng.to_string(index=False))
```

```python
# Import machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
# Prepare data for decision tree model
# We'll predict 'classification' based on department, position, and salary
data = df.copy()

# Encode categorical variables
le_dept = LabelEncoder()
le_pos = LabelEncoder()
le_class = LabelEncoder()

data['department_encoded'] = le_dept.fit_transform(data['department'])
data['position_encoded'] = le_pos.fit_transform(data['position'])
data['classification_encoded'] = le_class.fit_transform(data['classification'])

# Features and target
X = data[['department_encoded', 'position_encoded', 'salary']]
y = data['classification_encoded']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Data prepared for modeling.")
print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
```

```python
# Create and train decision tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

print("Decision tree model trained successfully.")
```

```python
# Evaluate the model
y_pred = dt_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le_class.classes_))

# Show predictions vs actual for test set
test_results = X_test.copy()
test_results['actual'] = le_class.inverse_transform(y_test)
test_results['predicted'] = le_class.inverse_transform(y_pred)
test_results['department'] = le_dept.inverse_transform(test_results['department_encoded'])
test_results['position'] = le_pos.inverse_transform(test_results['position_encoded'])
test_results = test_results[['department', 'position', 'salary', 'actual', 'predicted']]

print("\nTest Set Predictions:")
print(test_results)
```
